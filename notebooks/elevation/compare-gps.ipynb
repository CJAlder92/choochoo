```python
from ch2.data import *
from ch2.squeal import *

s = session('-v4')
st = activity_statistics(s, 'Raw Elevation', 'Elevation', 
                         'Altitude', 'Spherical Mercator X', 'Spherical Mercator Y', 'Distance',
                         time='2018-03-04 14:00')
st.describe()
```

```python
from bokeh.plotting import figure, output_notebook, show

output_notebook()
f = figure(plot_width=800, plot_height=250)

f.line(x='Distance', y='Altitude', source=st, color='black', line_width=1)
f.line(x='Distance', y='Raw Elevation', source=st, color='red', line_width=1)
f.line(x='Distance', y='Elevation', source=st, color='blue', line_width=1, line_dash='solid')
f.xaxis.axis_label = 'Distance'
f.yaxis.axis_label = 'Elevation / m'
show(f)
```

```python
from bokeh.layouts import row
from bokeh.models import HoverTool
from bokeh.tile_providers import CARTODBPOSITRON, STAMEN_TERRAIN, STAMEN_TONER

st2 = st.loc[(st['Distance'] > 55000) & (st['Distance'] < 59000)]
st2.describe()

ht = HoverTool(tooltips={'Distance': '@{Distance}'})

f = figure(plot_width=400, plot_height=400)
f.line(x='Distance', y='Altitude', source=st2, color='black', line_width=1)
f.line(x='Distance', y='Raw Elevation', source=st2, color='red', line_width=1)
f.line(x='Distance', y='Elevation', source=st2, color='blue', line_width=1, line_dash='solid')
f.xaxis.axis_label = 'Distance'
f.yaxis.axis_label = 'Elevation / m'

g = figure(plot_width=400, plot_height=400, x_axis_type='mercator', y_axis_type='mercator', tools=[ht])
g.line(x='Spherical Mercator X', y='Spherical Mercator Y', source=st2)
g.add_tile(CARTODBPOSITRON)

show(row(f, g))
```
