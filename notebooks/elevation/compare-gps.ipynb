```python
from ch2.data import *
from ch2.squeal import *

s = session('-v4')
st = activity_statistics(s, 'Raw Elevation', 'Elevation', 'Elevation 100', 'Elevation 300', 'Elevation 500',
                         'Altitude', 'Spherical Mercator X', 'Spherical Mercator Y', 'Distance',
                         time='2018-03-04 14:00')
print(st.columns)
print(st['Elevation'].describe())
st.describe()
```

```python
from bokeh.plotting import figure, output_notebook, show

output_notebook()
f = figure(plot_width=800, plot_height=250)
f.line(x='Distance', y='Raw Elevation', source=st, color='black')
f.line(x='Distance', y='Elevation 100', source=st, color='purple')
f.line(x='Distance', y='Elevation 300', source=st, color='blue')
f.line(x='Distance', y='Elevation 500', source=st, color='green')
f.line(x='Distance', y='Altitude', source=st, color='red')
f.xaxis.axis_label = 'Distance'
f.yaxis.axis_label = 'Elevation / m'
show(f)
```

```python
from bokeh.layouts import row
from bokeh.models import HoverTool
from bokeh.tile_providers import CARTODBPOSITRON, STAMEN_TERRAIN, STAMEN_TONER

st2 = st.loc[(st.index > '2018-03-04 14:10:00') & (st.index < '2018-03-04 14:12:00')]
st2.describe()

ht = HoverTool(tooltips={'Distance': '@{Distance}'})

f = figure(plot_width=400, plot_height=400, tools=[ht])
x = 'Distance'
f.line(x=x, y='Raw Elevation', source=st2, color='black')
f.line(x=x, y='Elevation 100', source=st2, color='purple')
f.line(x=x, y='Elevation 300', source=st2, color='blue')
f.line(x=x, y='Elevation 500', source=st2, color='green')
f.line(x=x, y='Altitude', source=st2, color='red')
f.xaxis.axis_label = x
f.yaxis.axis_label = 'Elevation / m'

g = figure(plot_width=400, plot_height=400, x_axis_type='mercator', y_axis_type='mercator', tools=[ht])
g.line(x='Spherical Mercator X', y='Spherical Mercator Y', source=st2)
g.add_tile(CARTODBPOSITRON)

show(row(f, g))
```
