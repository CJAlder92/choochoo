```python
from ch2.data import data
d = data('-v', '0')
stats = d.statistic_names()
#print(stats)
#list(zip((str(i) for i in stats.index.tolist()), stats['name'].tolist()))
print(stats.iloc[0])
```

```python
from bokeh.layouts import column, row
from bokeh.models.widgets import Dropdown, Button
from bokeh.io import show, output_notebook
from bokeh.plotting import figure

output_notebook()

def modify_doc(doc):
    
    f = figure(plot_width=800, plot_height=400)
    f.xaxis.axis_label = 'Date'
    m = Dropdown(label='Statistic', menu=list(zip(stats['name'].tolist(), (str(i) for i in stats.index.tolist()))))
    b = Button(label='Plot')
    r = row(column(m, b), f)
    
    def mkplot(index):
        f = figure(plot_width=500, plot_height=500)
        f.xaxis.axis_label = 'Date'
        r.children[1] = f
        row = stats.iloc[int(index)]
        name = row['name']
        data = d.statistic_journals(name, constraint=row['constraint'], owner=row['owner'])
        f.circle(data.index, data[name], line_width=2, legend=name)
        f.legend.location = 'top_left'    
        
    b.on_click(lambda: mkplot(m.value))
    doc.add_root(r)
    
show(modify_doc)
```
