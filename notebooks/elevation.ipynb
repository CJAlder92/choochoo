```python
from ch2.squeal.database import connect
from ch2.sortem.bilinear import bilinear_elevation_from_constant
from ch2.sortem.spline import spline_elevation_from_constant

_, log, db = connect(['-v5'])
s = db.session()
bilinear = bilinear_elevation_from_constant(log, s)
spline0 = spline_elevation_from_constant(log, s, smooth=0)
spline1 = spline_elevation_from_constant(log, s, smooth=1)
spline3 = spline_elevation_from_constant(log, s, smooth=3)
spline10 = spline_elevation_from_constant(log, s, smooth=10)
print(bilinear.elevation(-33.436, -70.607))
print(spline0.elevation(-33.431, -70.588))
```

```python
import numpy as np

x = np.linspace(-70.607, -70.588, 300)
y = np.linspace(-33.436, -33.431, 300)
zb = [bilinear.elevation(Y, X) for X, Y in zip(x, y)]
zs0 = [spline0.elevation(Y, X) for X, Y in zip(x, y)]
zs1 = [spline1.elevation(Y, X) for X, Y in zip(x, y)]
zs3 = [spline3.elevation(Y, X) for X, Y in zip(x, y)]
zs10 = [spline10.elevation(Y, X) for X, Y in zip(x, y)]
```

```python
from bokeh.plotting import figure, output_notebook, show

output_notebook()
f = figure(plot_width=800, plot_height=400, title='Elevation along Pocuro')
f.line(x=x, y=zb, line_color='red', legend='bilinear')
f.line(x=x, y=zs0, line_color='green', legend='spline')
f.line(x=x, y=zs1, line_color='blue', legend='spline 1')
f.line(x=x, y=zs3, line_color='purple', legend='spline 3')
f.line(x=x, y=zs10, line_color='black', legend='spline 10')
f.legend.location = 'top_left'
show(f)
```
