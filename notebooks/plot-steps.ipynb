```python
from ch2.data import *
s = session()
d = statistics(s, 'Daily Steps')
```

```python
import datetime as dt
from bokeh.plotting import figure, output_notebook, show
from ch2.lib.date import to_time
def plot(data, key):
    output_notebook()
    f = figure(plot_width=800, plot_height=250, x_axis_type='datetime', title='Steps')
    f.vbar(x=data.index, width=dt.timedelta(hours=20), top=data[key])
    f.xaxis.axis_label = 'Date'
    f.yaxis.axis_label = 'Steps / day'
    show(f)
plot(d[to_time('2018-09-15'):], 'Daily Steps')
```

```python
d['m5'] = d['Daily Steps'].rolling(5).mean(center=True)
plot(d[to_time('2018-09-15'):], 'm5')
```

```python
output_notebook()
d['m10'] = d['Daily Steps'].rolling(10).mean(center=True)
data = d[to_time('2018-09-15'):]
f = figure(plot_width=800, plot_height=250, x_axis_type='datetime', title='Steps')
f.vbar(x=data.index, width=dt.timedelta(hours=20), top=data['Daily Steps'])
f.line(x=data.index, y=data['m5'], color='orange', line_width=2, line_dash='dashed')
f.line(x=data.index, y=data['m10'], color='firebrick', line_width=3)
f.xaxis.axis_label = 'Date'
f.yaxis.axis_label = 'Steps / day'
show(f)
```
