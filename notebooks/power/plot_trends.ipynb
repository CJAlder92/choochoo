```python
import pandas as pd
from ch2.data import *
from ch2.squeal import *

s = session('-v 5')
ad = s.query(StatisticName).filter(StatisticName.name.like('%Active Distance%')).all()
print([x.name for x in ad])
mad_stats = statistics(s, 'Sum/Month Active Distance')
print(mad_stats.describe())
pwr_stats = statistics(s, POWER_HR, POWER_HR_LAG)
print(pwr_stats.describe())
```

```python
from bokeh.plotting import output_notebook, show
from ch2.data.plot import *

output_notebook()
f = figure(plot_width=800, plot_height=450, x_axis_type='datetime', y_range=(0, 2e6))
f.circle(x=mad_stats.index, y=mad_stats['Sum/Month Active Distance'], size=10, fill_color='red')
f.circle(x=pwr_stats.index, y=pwr_stats[POWER_HR] * 10, fill_color='green')
f.circle(x=pwr_stats.index, y=pwr_stats[POWER_HR_LAG] * 1e4, fill_color='yellow')
show(f)
```
