```python
import pandas as pd
from ch2.uranus.route.power import *
from ch2.uranus.route.read import *
from ch2.data import *
from ch2.squeal.database import connect

s = session('-v 0')
activity = std_route(s, step='20s', time='2018-03-01 16:00', activity_group_name='Bike')
add_energy_budget(activity, 64+12)
add_loss_estimate(activity)
add_power_estimate(activity)
activity.describe()
```

```python
from bokeh.plotting import output_notebook, show
from bokeh.layouts import row, column

from ch2.data.plot import *

activity['P2'] = activity[POWER].rolling('300s', min_periods=1).median()

output_notebook()

sp = comparison_line_plot(700, 200, DISTANCE, SPEED, activity, ylo=0, x_range=el.x_range)
sp_c = cumulative_plot(200, 200, SPEED, activity, ylo=0)

pw = comparison_line_plot(700, 200, DISTANCE, 'P2', activity, ylo=0, yhi=200, x_range=el.x_range)
pw_c = cumulative_plot(200, 200, POWER, activity, ylo=0)

show(column(row(sp, sp_c), row(pw, pw_c)))
```
