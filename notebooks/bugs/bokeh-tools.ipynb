```python
import pandas as pd
import numpy as np

def make_data(n_points, n_columns):
    return pd.DataFrame(np.random.randn(n_points, n_columns))
    
make_data(100, 3).describe()
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.layouts import column, row
from bokeh.models import PanTool, ZoomInTool, HoverTool, ZoomOutTool, ResetTool, Range1d

def make_plots(df):
    x_range = None
    for col in df:
        y_range = Range1d(start=df[col].min(), end=df[col].max())
        f = figure(plot_width=600, plot_height=200, x_range=x_range, y_range=y_range,
                  tools=[PanTool(dimensions='width'), ZoomInTool(dimensions='width'), 
                         ZoomOutTool(dimensions='width'), ResetTool()])
        f.toolbar.logo = None
        f.line(x=df.index, y=df[col])
        if x_range is None:
            x_range = f.x_range
        g = figure(plot_width=200, plot_height=200)
        g.line(x=df.index, y=df[col])
        g.toolbar_location = None
        yield row([f, g])
```

```python
output_notebook()
show(column(list(make_plots(make_data(10000, 3)))))
```
