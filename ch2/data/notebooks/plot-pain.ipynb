```python
from ch2.data import data
from bokeh.plotting import figure, output_notebook, show
ch2 = data('-v', '0')
print(ch2.injury_names())
data = ch2.injuries('Tendon pain.*')['Tendon pain (femur, lhs)']
output_notebook()
f = figure(plot_width=800, plot_height=250, x_axis_type='datetime', title='Average Pain')
f.circle(x=data.index, y=data['pain average'])
f.xaxis.axis_label = 'Date'
f.yaxis.axis_label = 'Pain / 0-9'
show(f)
```
