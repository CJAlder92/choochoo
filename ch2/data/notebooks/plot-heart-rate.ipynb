```python
from ch2.data import data
from bokeh.plotting import figure, output_notebook, show
ch2 = data('-v', '0')
data = ch2.diaries()
output_notebook()
f = figure(plot_width=800, plot_height=250, x_axis_type='datetime', title='Rest HR')
f.circle(x=data.index, y=data['rest heart rate'])
f.xaxis.axis_label = 'Date'
f.yaxis.axis_label = 'Heart Rate / bpm'
show(f)
```
