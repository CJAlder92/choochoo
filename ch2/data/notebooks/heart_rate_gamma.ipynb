```python
def mk_zones(min_zone, max_zone, resolution=100):
    for i in range(int((max_zone - min_zone) * resolution)):
        yield min_zone + i / resolution
            
zones = list(mk_zones(1.5, 6.5))
```

```python
def mk_impulse(zone, gamma=1):
    return ((max(zone, 2) - 2) / 4) ** gamma

impulses = [mk_impulse(z) for z in zones]
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import Dark2
output_notebook()
f = figure(plot_width=400, plot_height=400, title='How gamma Affects Impulse')
f.xaxis.axis_label = 'Heart Rate Zone'
f.yaxis.axis_label = 'Impulse per second'
for i, gamma in enumerate((0.5, 1, 2)):
    f.line(zones, [mk_impulse(z, gamma=gamma) for z in zones], 
           line_width=2, color=Dark2[3][i], legend='gamma %s' % gamma)
f.legend.location = 'top_left'    
show(f)
```
