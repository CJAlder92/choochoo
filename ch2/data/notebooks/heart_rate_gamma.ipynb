```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import Dark2
from ch2.data import *

RESTHR, MAXHR, FTHR, KM, KF = 50, 170, 154, 1.92, 1.67
hr = list(range(RESTHR, MAXHR))

output_notebook()
f = figure(plot_width=500, plot_height=500, title='How gamma Affects Impulse')
f.xaxis.axis_label = 'Heart Rate'
f.yaxis.axis_label = 'Impulse per second'
pt = Dark2[3]
for i, gamma in enumerate([0.5, 1, 2]):
    f.line([zone(x, FTHR) for x in hr], [shrimp(x, gamma, 0, FTHR) for x in hr], 
           line_width=2, color=pt[i], legend='gamma %s' % gamma)
f.legend.location = 'top_left'    
show(f)
```
