```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.tile_providers import CARTODBPOSITRON, STAMEN_TERRAIN, STAMEN_TONER
frames = d.activity_waypoints('Bike', '2018-09-06')
#print(frames[0])
output_notebook()
f = figure(plot_width=800, plot_height=800, x_axis_type='mercator', y_axis_type='mercator')
for frame in frames:
    f.circle(x='x', y='y', source=frame)
f.add_tile(CARTODBPOSITRON)
show(f)
```
