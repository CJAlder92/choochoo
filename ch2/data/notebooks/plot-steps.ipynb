```python
from ch2.data import data
d = data()
d.statistics('Steps')
```

```python
sj = d.statistic_journals('Steps')
sj.describe()
```

```python
from ch2.lib.date import to_time
recent = sj[to_time('2018-09-10'):]
```

```python
import datetime as dt
from bokeh.plotting import figure, output_notebook, show
output_notebook()
f = figure(plot_width=800, plot_height=250, x_axis_type='datetime', title='Steps')
f.vbar(x=recent.index, width=dt.timedelta(hours=20), top=recent['Steps'])
f.xaxis.axis_label = 'Date'
f.yaxis.axis_label = 'Steps / day'
show(f)
```
