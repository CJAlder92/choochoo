```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import Paired
from ch2.data import *

output_notebook()
f = figure(plot_width=500, plot_height=500, title='Comparing Impulse Models')
f.xaxis.axis_label = 'Heart Rate'
f.yaxis.axis_label = 'Impulse per Second'

RESTHR, MAXHR, FTHR, KM, KF = 50, 170, 154, 1.92, 1.67
hr = list(range(RESTHR, MAXHR))
trm = [trimp(x, RESTHR, MAXHR, KM) for x in hr]
trf = [trimp(x, RESTHR, MAXHR, KF) for x in hr]
ed = [edwards(x, MAXHR) for x in hr]
sh10 = [shrimp(x, 1, 0, FTHR) for x in hr]
sh127 = [0.7 * shrimp(x, 1, 2, FTHR) for x in hr]
sh207 = [0.7 * shrimp(x, 2, 0, FTHR) for x in hr]
sh206 = [0.6 * shrimp(x, 2, 0, FTHR) for x in hr]

pt = Paired[7]
f.line(hr, sh10, line_width=2, color=pt[0], legend='SHRIMP 1,0')
f.line(hr, sh127, line_width=2, color=pt[1], legend='SHRIMP 1,2 x 0.7')
f.line(hr, sh207, line_width=2, color=pt[2], legend='SHRIMP 2,0 x 0.7')
f.line(hr, sh206, line_width=2, color=pt[3], legend='SHRIMP 2,0 x 0.6')
f.line(hr, trm, line_width=2, color=pt[4], legend='TRIMP M')
f.line(hr, trf, line_width=2, color=pt[5], legend='TRIMP F')
f.line(hr, ed, line_width=2, color=pt[6], legend='Edwards')

f.legend.location = 'top_left'    
show(f)
```
