```python
from ch2.data import data
ch2 = data('-v', '0')
cycling = ch2.activity('Cycling')
stats = cycling.statistics('.*')
for s in stats:
    print(s.name, ':', s.units)
```

```python
from bokeh.plotting import figure, output_notebook, show
from ch2.data.plot import boxplot

data = cycling.summary_statistics('Max med HR over 30m')

output_notebook()
f = figure(plot_width=800, plot_height=250, x_axis_type='datetime')
boxplot(f, data['Max med HR over 30m'])
show(f)
```
