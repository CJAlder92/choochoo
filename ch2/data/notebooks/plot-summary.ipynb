```python
from bokeh.plotting import figure, output_notebook, show
from ch2.data import bokeh_boxplot
data = d.statistic_quartiles('Max Med HR over 30m')
output_notebook()
f = figure(plot_width=800, plot_height=250, x_axis_type='datetime', title='Max median HR over 30 min')
bokeh_boxplot(f, data['Max Med HR over 30m'])
f.xaxis.axis_label = 'Date'
f.yaxis.axis_label = 'Heart rate / bpm'
show(f)
```
